```python
file = 'list.txt'

with open(file, 'r') as f:
    wordlist = f.readlines()
```

```python
print(wordlist)
```

```python
char = "0123456789"
```

```python
wordlist = [word.strip() for word in wordlist]
```

```python
wordlist = [''.join(filter(lambda x: x in char, word)) for word in wordlist]
```

```python
finallist = []

for numbers in wordlist:
    # add the first number to the last number and remove the between numbers
    if len(numbers) > 1:
        result = (numbers[0]) + (numbers[-1])
        print(f'{result}')
        finallist.append(result)
    else:
        finallist.append(str(numbers) + str(numbers))
        
    
```

```python
print(finallist)
```

```python
total = 0
for i in finallist:
    total += int(i)
```
