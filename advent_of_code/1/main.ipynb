```python
file = 'list.txt'

with open(file, 'r') as f:
    wordlist = f.readlines()
```

```python
print(wordlist)
```

```python
char = "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ"
```

```python
wordlist = [word.strip() for word in wordlist]
```

```python
# remove chars and let numbers only
wordlist = [''.join(filter(lambda x: x in char, word)) for word in wordlist]
```
